`brew services start postgresql`でpostgresqlを開始

`brew services stop postgresql`でpostgresqlを停止

```python
import pandas as pd
import psycopg2
```

```python
conn = psycopg2.connect(
    dbname="main-app",
)
df_answers = pd.read_sql_query("SELECT * FROM main_app_answers", conn, index_col="id")
df_kinds = pd.read_sql_query("SELECT * FROM main_app_kinds", conn)
df_modelnames = pd.read_sql_query("SELECT * FROM main_app_modelnames", conn)
df_respondents = pd.read_sql_query("SELECT * FROM main_app_respondents", conn)
df_samplenames = pd.read_sql_query("SELECT * FROM main_app_samplenames", conn)
df_speakernames = pd.read_sql_query("SELECT * FROM main_app_speakernames", conn)
conn.close()

df_kinds = df_kinds.rename(columns={"id": "kind_id"})
df_modelnames = df_modelnames.rename(columns={"id": "model_name_id"})
df_respondents = df_respondents.rename(columns={"id": "respondent_id"})
df_samplenames = df_samplenames.rename(columns={"id": "sample_name_id"})
df_speakernames = df_speakernames.rename(columns={"id": "speaker_name_id"})

df_answers = df_answers.merge(df_kinds, on=["kind_id"], how="inner")
df_answers = df_answers.merge(df_modelnames, on=["model_name_id"], how="inner")
df_answers = df_answers.merge(df_respondents, on=["respondent_id"], how="inner")
df_answers = df_answers.merge(df_samplenames, on=["sample_name_id"], how="inner")
df_answers = df_answers.merge(df_speakernames, on=["speaker_name_id"], how="inner")
```
